# Simulating sensor data

```python
nSamples = 500
inputRange = np.linspace(0,6,nSamples)
signalTruth = np.sin(inputRange)

noiseVar = np.random.normal(loc = 0, scale = 1.0,size = nSamples)

signal = noiseVar/5 + signalTruth
```

```python
plt.plot(inputRange,signal, 'b.', label = "Truth")
plt.plot(inputRange,signalTruth, 'r-', linewidth = 4, label = "Sensor readings")
plt.legend()
plt.xlabel("Some independent value")
plt.ylabel("Some dependent value")
plt.show()
```

```python
from scipy.signal import savgol_filter
```

```python
filteredSignal = savgol_filter(signal, window_length = 101, polyorder=5)
```

# Filtering noise

```python
plt.plot(inputRange,signal, 'b.', label = "Sensor readings")
plt.plot(inputRange,filteredSignal, 'r--.', label = "Reconstructed signal")
plt.xlabel("Some independent value")
plt.ylabel("Some dependent value")
plt.show()
```
